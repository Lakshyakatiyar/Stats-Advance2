4.Importance of Normal Distribution:

The normal distribution is fundamental in statistics because it describes many natural phenomena.
Many statistical tests and procedures assume normality due to its mathematical properties.
The Central Limit Theorem states that the distribution of the sample mean approaches a normal distribution as the sample size grows, regardless of the original distribution.
Real-life Examples:

IQ Scores: Intelligence quotient scores are normally distributed with a mean of 100 and a standard deviation of 15.
Blood Pressure: Blood pressure readings in a healthy population typically follow a normal distribution.
Stock Prices: Daily returns of stock prices often follow a normal distribution.
